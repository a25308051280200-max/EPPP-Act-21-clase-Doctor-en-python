```python
# Definición de la Clase
class Deportista:
    def __init__(self, compromiso, inteligencia, nombre):
        # Atributos (Datos referentes a la clase)
        self.compromiso = compromiso    # Atributo 1
        self.inteligencia = inteligencia # Atributo 2
        self.nombre = nombre            # Atributo 3

    # --- Funciones (Acciones que realiza la Deportista) ---

    def correr(self):
        print(f"{self.nombre} está corriendo en la pista de atletismo.")

    def calentar(self):
        print(f"{self.nombre} está calentando los músculos. Nivel de inteligencia aplicada: {self.inteligencia}.")

    def entrenamientos(self):
        print(f"{self.nombre} ha iniciado su sesión de entrenamientos. Compromiso: {self.compromiso}.")

# --- Creación de Objetos (Instancias) ---

# Deportista1 con datos de ejemplo
deportista1 = Deportista("Alto", "Estratégica", "Ana Pérez")

# Deportista2 con datos de ejemplo
deportista2 = Deportista("Máximo", "Analítica", "Sofía García")

# --- Llamadas a las funciones para cada objeto ---

print("Reporte Deportista 1:")
deportista1.calentar()
deportista1.correr()
deportista1.entrenamientos()

print("\nReporte Deportista 2:")
deportista2.calentar()
deportista2.correr()
deportista2.entrenamientos()

print("\n--- Programa terminado ---")
print("Creador de software por Alejandra Rivera Valenzuela NL = 45")
```
